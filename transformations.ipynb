```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('./figures/parking_lot.png')

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply morphological operations
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
morphed = cv2.morphologyEx(gray, cv2.MORPH_CLOSE, kernel)

# Apply Gaussian blur to reduce noise
blurred = cv2.GaussianBlur(morphed, (5, 5), 0)

# Apply Canny edge detection
edges = cv2.Canny(blurred, 50, 150)

# Find contours
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on the original image
result = image.copy()
cv2.drawContours(result, contours, -1, (0, 255, 0), 2)

# Create a subplot to show each step
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()

axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(gray, cmap='gray')
axes[1].set_title('Grayscale Image')
axes[1].axis('off')

axes[2].imshow(morphed, cmap='gray')
axes[2].set_title('Morphological Operations')
axes[2].axis('off')

axes[3].imshow(blurred, cmap='gray')
axes[3].set_title('Gaussian Blur')
axes[3].axis('off')

axes[4].imshow(edges, cmap='gray')
axes[4].set_title('Canny Edges')
axes[4].axis('off')

axes[5].imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
axes[5].set_title('Contours')
axes[5].axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('./figures/parking_lot.png')

# Convert the image to HSV color space
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define the range for the green color
lower_green = np.array([35, 40, 40])
upper_green = np.array([85, 255, 255])

# Create a mask to filter out the green color
mask = cv2.inRange(hsv, lower_green, upper_green)

# Invert the mask to keep everything except the green color
mask_inv = cv2.bitwise_not(mask)

# Apply the mask to the original image
result = cv2.bitwise_and(image, image, mask=mask_inv)

# Convert the result to grayscale
gray = cv2.cvtColor(result, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to reduce noise
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Apply Canny edge detection
edges = cv2.Canny(blurred, 50, 150)

# Find contours
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on the original image
contour_result = image.copy()
cv2.drawContours(contour_result, contours, -1, (0, 255, 0), 2)

# Create a subplot to show each step
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.ravel()

axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB))
axes[1].set_title('HSV Image')
axes[1].axis('off')

axes[2].imshow(mask, cmap='gray')
axes[2].set_title('Green Mask')
axes[2].axis('off')

axes[3].imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
axes[3].set_title('Result without Green')
axes[3].axis('off')

axes[4].imshow(gray, cmap='gray')
axes[4].set_title('Grayscale Image')
axes[4].axis('off')

axes[5].imshow(blurred, cmap='gray')
axes[5].set_title('Gaussian Blur')
axes[5].axis('off')

axes[6].imshow(edges, cmap='gray')
axes[6].set_title('Canny Edges')
axes[6].axis('off')

axes[7].imshow(cv2.cvtColor(contour_result, cv2.COLOR_BGR2RGB))
axes[7].set_title('Contours')
axes[7].axis('off')

plt.tight_layout()
plt.show()
```
